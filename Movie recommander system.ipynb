```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits , on ='title')
movies.shape
```

```python
movies.head(1)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies= movies[['genres','id','keywords','title','overview','cast','crew']]
movies.shape
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# to change into a formet like ['A','B','C']
import json
def convertList(obj):
    A=[];
    for items in json.loads(obj):
       A.append(items['name'])
    return A;
```

```python
movies['genres']=movies['genres'].apply(convertList)
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convertList)
movies.head(1)
```

```python
def convertList3(obj):
    A=[];
    counter=0;
    for items in json.loads(obj):
           if(counter!=3):
             A.append(items['name'])
             counter+=1;
           else:
             break
       
    return A;
```

```python
movies['cast']=movies['cast'].apply(convertList3)
movies.head(2)
```

```python
def getDirector(obj):
    A=[];
    for items in json.loads(obj):
           if(items['job']=='Director'):
             A.append(items['name'])
             break
    return A;
```

```python
movies['crew']=movies['crew'].apply(getDirector)
movies.head(3)
```

```python
movies.overview[0]
```

```python
movies['overview']=movies['overview'].apply( lambda x : x.split(' '))
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew'] 
```

```python
movies.head(1)
```

```python
new_df = movies.drop(['genres','keywords','overview','cast','crew'] ,axis=1)
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

```python
# there can be many same works so i am applying stemming to my tags
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split(' '):
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: stem(x))
new_df.tags[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# using beg of words algorithm
cv = CountVectorizer(max_features=5000, stop_words='english')
# converting into nu,py array
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_ind = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_ind]
    movie_list = sorted(list(enumerate(distance)),reverse = True, key = lambda X : X[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
